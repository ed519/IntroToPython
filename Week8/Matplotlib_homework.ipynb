### P1. Write a Python code to output the following graph. (or a similar one)
```
The productivity values for the 3 groups are the following: 

g1 = [80, 60, 80, 67, 34]
g2 = [45, 67, 23, 89, 10]
g3 = [30, 90, 50, 20, 90]
```

````python
import numpy as np
````

````python
index = np.arange(5)

g1 = [80, 60, 80, 67, 34]
g2 = [45, 67, 23, 89, 10]
g3 = [30, 90, 50, 20, 90]
days = ['M','T','W','Th','F']
bar_width = 0.25

plt.bar(index, g1, bar_width, color='r', label='Group1')
plt.bar(index+bar_width, g2 , bar_width,color='c', label='Group2')
plt.bar(index+(2*bar_width), g3 , bar_width,color='y', label='Group3')



plt.xlabel("Groups")
plt.ylabel("Productivity")
plt.title("Productivity during the week")

plt.legend()
plt.show()
````
